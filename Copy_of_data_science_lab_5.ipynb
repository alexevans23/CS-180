## Exercise 1: Protein Powder
Carbs, fats and proteins are the three primary macro nutrients. Create a figure plotting the distribution of each of these macro nutrients together (i.e., three distributions on a single plot). Make sure to provide a legend.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
cerealData = pd.read_csv("cereal.csv") 

cerealData['carbo'].plot.hist(alpha = 0.5, label = 'carbs')
cerealData['protein'].plot.hist(alpha = 0.5, label = 'proteins')
cerealData['fat'].plot.hist(alpha = 0.5, label = 'fats')

plt.legend()

plt.show()

```

## Exercise 2: Sugar Daddy
Get a list of the top 5 most sugary cereals and the 5 least sugary cereals.

```python
top_sugar = cerealData.sort_values(by = 'sugars', ascending = False)
sugariest = top_sugar.head(5)
sugariest[['name', 'sugars']]
```

```python
least_sugar = cerealData.sort_values(by = 'sugars', ascending = True)
healthiest = least_sugar.head(5)
healthiest[['name', 'sugars']]
```

## Exercise 3: Cereal Killer
Get a list of the top 5 highest rated and lowest rated cereals.

```python
rated_cereals = cerealData.sort_values(by = 'rating', ascending = False)
top_rated = rated_cereals.head(5)
top_rated[['name','rating']]
```

```python
lowest_cereals = cerealData.sort_values(by = 'rating', ascending = True)
lowest_rated = lowest_cereals.head(5)
lowest_rated[['name', 'rating']] 
```

## Exercise 4: America
Quantify the relationship between sugar and ratings. 

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship. It may be helpful to look at the [seaborn regplot documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html").

```python
sns.regplot(x = 'sugars', y = 'rating', data = cerealData)
plt.xlabel("Sugar (g)")
plt.ylabel("Ratings")
plt.title("Relationship Between Sugar and Ratings")
plt.show()
```


Calculate a correlation statistic describing the relationship between sugar and ratings  (i.e., r or r squared).

```python
# Calculate the statistic using this cell:
r = cerealData['sugars'].corr(cerealData['rating'])
r ** 2
```

Write a statement in plain English interpreting this statistic.

This shows us the proportion of how the amount of sugar varies based on the rating of the cereal. 1 is the biggest number so .577 is pretty correlated.


## Exercise 5: America Part 2
Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
sns.regplot(x = "carbo", y = "rating", data = cerealData)
plt.xlabel("Carbs")
plt.ylabel("Rating")
plt.title("Relationship between Carbs and Rating")
plt.show()

sns.regplot(x = "sugars", y = "rating", data = cerealData)
plt.xlabel("Sugar")
plt.ylabel("Rating")
plt.title("Relationship between Sugars and Rating")
plt.show()

sns.regplot(x = "calories", y = "rating", data = cerealData)
plt.xlabel("Calories")
plt.ylabel("Rating")
plt.title("Relationship between Calories and Rating")
plt.show()

sns.regplot(x = "protein", y = "rating", data = cerealData)
plt.xlabel("Protein")
plt.ylabel("Rating")
plt.title("Relationship between Protein and Rating")
plt.show()

sns.regplot(x = "fat", y = "rating", data = cerealData)
plt.xlabel("Fat")
plt.ylabel("Rating")
plt.title("Relationship between Fats and Rating")
plt.show()

```

```python
r = cerealData['calories'].corr(cerealData['rating'])
r ** 2
```

Of the variables carbo, sugars, calories, protein, and fat, which has the strongest relationship with rating? Justify your answer.

The strongest relationship is sugar's with ratings. The closest two are sugar and calories, but as I showed above, the sugar has a larger r-squared.
